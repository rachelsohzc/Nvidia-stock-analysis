## Nvidia stock analysis

I've been interested in investing in the stock market lately. I thought doing this project might help me learn more about valuing companies and deciding what stocks to invest in. This is a short project where I analyse Nvidia stocks using past information. Some key questions that I wanted to answer in this project are:

1. What was the change in price of stock over time? 
2. What was the daily and monthly return of the stock?
3. What is the stock's moving average? 
4. How much risk do we take by investing in this stock? 
5. How can we predict future stock behavior with statistical models?

The analysis for this can be found below.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'NVidia_stock_history.csv'
nvidia = pd.read_csv(url)
len(nvidia.index)
```

```python
#Deleting null records
nvidia = nvidia.dropna(how='any',axis=0)
len(nvidia.index)
```

```python
nvidia = nvidia.sort_values('Date')
nvidia.head()
```

#### Stock features

For this analysis, we can mainly focus on the closing price on each day.

```python
#Seeing how the price changes over time
plt.figure(figsize = (20,9))
plt.plot(nvidia.Date, nvidia.Close)
plt.xticks(range(0,nvidia.shape[0],100),nvidia['Date'].loc[::100],rotation=45)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Mid price',fontsize=12)
plt.title('Nvidia stock price changes over time',fontsize=20)
plt.show()
```

```python
#Volume of stocks traded over time
plt.figure(figsize = (20,9))
plt.plot(nvidia.Date, nvidia.Volume)
plt.xticks(range(0,nvidia.shape[0],100),nvidia['Date'].loc[::100],rotation=45)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Volume',fontsize=12)
plt.title('Volume of Nvidia stocks traded over time',fontsize=20)
plt.show()
```

```python
#The daily return 
nvidia['Daily_returns'] = nvidia.Close.pct_change()
nvidia.head()
```

```python
plt.figure(figsize=(20,9))
plt.plot(nvidia.Date, nvidia.Daily_returns)
plt.xticks(range(0,nvidia.shape[0],100),nvidia.Date.loc[::100],rotation=45)
plt.title('Daily returns of stock',fontsize=18)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Daily return',fontsize=12)
plt.show()
```

Often, in technical analysis, we tend to use short-term and long-term moving averages to assess the price pattern in the future. Moving averages are calculated based on subsets of the data. For instance, for the short-term moving average, we can take subsets of 20 days over a year. This would lead to 18 data points across the year. Moving averages are mainly used to remove noise in the data.

```python
#Short-term moving average of 20 days
short_MA = nvidia.rolling(window=20).mean()
short_MA.head()
```

```python
#Long-term moving average of 80 days
long_MA = nvidia.rolling(window=80).mean()
long_MA.head()
```

```python
#Plotting the price, long MA and short MA for comparison
plt.figure(figsize=(20,9))

plt.xlabel('Dates',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.title('Moving average in comparison to price',fontsize=18)

plt.plot(nvidia.Date,nvidia.Close,label='Price')
plt.plot(nvidia.Date,short_MA.Close, label='Short MA')
plt.plot(nvidia.Date,long_MA.Close,label='Long MA')

plt.xticks(range(0,nvidia.shape[0],100),nvidia.Date.loc[::100],rotation=45)

plt.legend()
plt.show()
```

The trends in price changes are only interesting after 2015-05-11. Prior to this date, the fluctuations are minimal so we can take out this subset of the data.

```python
#Re-plotting the data because the price changes after 2015-05-11 is more interesting
nvidia['Date'] = pd.to_datetime(nvidia['Date'])

mask = (nvidia.Date >= '11/05/2015') & (nvidia.Date <= '16/09/2021')
nvidia[mask].head(10)
```

```python
modifiednvidia = nvidia.loc[mask]
modifiedshortMA = short_MA.loc[mask]
modifiedlongMA = long_MA.loc[mask]
```

```python
plt.figure(figsize=(20,9))

plt.xlabel('Dates',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.title('Moving average in comparison to price',fontsize=18)

plt.plot(modifiednvidia.Date, modifiednvidia.Close, label='Price')
plt.plot(modifiednvidia.Date, modifiedshortMA.Close, label='Short MA')
plt.plot(modifiednvidia.Date, modifiedlongMA.Close, label='Long MA')

plt.legend()
plt.show()
```

```python
#We calculate risk by comparing the expected return to the standard deviation of daily returns
nvidiarisk = modifiednvidia.Close
nvidiarisk.head()
```

```python
plt.figure(figsize=(20,9))

plt.xlabel('Expected return',fontsize=12)
plt.ylabel('Risk',fontsize=12)
plt.title('Risk of Nvidia stock',fontsize=18)

plt.scatter(nvidiarisk.mean(),nvidiarisk.std())

plt.show()
```

#### Stock price predictions

There are a number of methods used to predict stock prices. But for this project in particular, we'll look at some time series models. Artificial intelligence models are also used but I will attempt that in another project. Here, my main focus will be building the AutoRegressive Integrated Moving Average (ARIMA) model. The ARIMA model is a time series model that takes autocorrelation into account to predict future values.

If you want to understand ARIMA models from a more theoretical perspective, refer [here](https://www.projectpro.io/article/how-to-build-arima-model-in-python/544#:~:text=Model%20in%20Python%3F-,ARIMA%20Model%2D%20Complete%20Guide%20to%20Time%20Series%20Forecasting%20in%20Python,data%20to%20predict%20future%20values.).

```python
from numpy import log
from pandas.plotting import lag_plot
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
```

```python
#Finding the order of differencing for the stationary time series
#If d=0, it means the time series is stationary and vice versa
result = adfuller(nvidia.Close)
print('The ADF Statistic: %f' % result[0])
print('The p-value: %f' % result[1])
```

```python
plot_acf(nvidia.Close)
```

```python
#Testing for cross-correlation
plt.figure()
lag_plot(nvidia['Close'], lag = 3)
plt.title('Nvidia Stock - Autocorrelation plot with lag = 3')
plt.show()
```

```python
#Splitting the data into training and testing
from sklearn.model_selection import train_test_split
training, testing = train_test_split(nvidia, test_size = 0.3, random_state = 21)

train_data = training['Close'].values
test_data = testing['Close'].values
```

```python
#Building the model
history = [x for x in train_data]
model_preds = []
N_test_obs = len(test_data)

for time_point in range(N_test_obs):
    model = sm.tsa.arima.ARIMA(history, order = (4,1,0))
    model_fit = model.fit()
    out = model_fit.forecast()
    y_pred = out[0]
    model_preds.append(y_pred)
    y_orig = test_data[time_point]
    history.append(y_orig)
```

```python
#Measuring accuracy of the ARIMA model
MSE_error = mean_squared_error(test_data, model_preds)
print('Testing MSE is {}'.format(MSE_error))
```
